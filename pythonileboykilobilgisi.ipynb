```python
print(veri_seti.columns) #degiskenler
```

```python
boy_bilgisi = veri_seti['BOY']
print(boy_bilgisi)
```

```python
kilo_bilgisi = veri_seti['KİLO']
print(kilo_bilgisi)
```

```python
boykilo_bilgisi = veri_seti[['BOY', 'KİLO']]
```

```python
boykilo_bilgisi.columns = ['boy(cm)', 'kilo(kg)']
print(boykilo_bilgisi)
```

```python
plt.plot(boykilo_bilgisi)#grafikleştirme işlemi başladı
plt.legend(('boy(cm)', 'kilo(kg)'), loc='upper left')#veri tanımlaması
plt.title('veri tablosu-boy&kilo')#grafik başlık bilgisi
plt.ylabel('y ekseni-veri ölçüm değeri')#y ekseni tanımlaması
plt.xlabel('x ekseni-kişi sıra(ID)bilgisi')#x ekseni tanımlaması
plt.figure(figsize=(10,5))#grafiğin çıktı boyutumu ayarlanması
plt.show()#çizilen grafiğin gösterimi
```
